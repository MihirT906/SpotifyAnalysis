### Feature Selection

```python
import statsmodels.api as sm
def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features, dtype='float64')
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break 
        else:
            break
    return best_features
```

```python
#best_columns = stepwise_selection(X,y)

#Manually chose columns to make it easier
best_columns = ['explicit', 'energy', 'loudness', 'speechiness', 'instrumentalness']
```

```python
X = X[best_columns]
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Multiple Regression

```python
X_train
```

```python
y_train = pd.DataFrame(y_train, columns = ['popularity'])
y_test = pd.DataFrame(y_test, columns = ['popularity'])
y_train
```

### Training the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
y_pred
#np.concatenate((y_pred, y_test), axis=1)
```

### Testing the model

```python
y_test
```

```python
prediction_df = pd.DataFrame(columns = ['predicted value'], data = y_pred)
prediction_df['actual value'] = np.array(y_test)
prediction_df
```

### Creating a Summary Table

```python
feature_names = X.columns.values
feature_names
```

```python
summary_table = pd.DataFrame(columns = ['Feature name'], data = feature_names)
summary_table['Coefficient'] = np.transpose(regressor.coef_)
```

```python
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept', regressor.intercept_]
```

```python
summary_table = summary_table.sort_values('Coefficient', ascending=False, key=abs)
summary_table
```

### Evaluate Model

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
plt.scatter(y_test, y_pred, c ="blue")
x = np.linspace(0,100,100)
plt.plot(x,x)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

# SVM Regression

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_columns = X_train.columns.values
```

```python
X_train = sc_X.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns=X_columns)
X_train
```

```python
X_test = sc_X.transform(X_test)
X_test = pd.DataFrame(X_test, columns=X_columns)
X_test
```

```python
y_train = sc_y.fit_transform(y_train)
y_train = pd.DataFrame(y_train, columns=['popularity'])
y_train
```

```python
y_test = sc_y.transform(y_test)
y_test = pd.DataFrame(y_test, columns=['popularity'])
y_test
```

### Training the Model

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(X_test))
y_pred
```

```python
prediction_df = pd.DataFrame(columns = ['predicted value'], data = y_pred)
prediction_df['actual value'] = np.array(sc_y.inverse_transform(y_test))
prediction_df
```

```python
plt.scatter(sc_y.inverse_transform(y_test), y_pred, c ="blue")
x = np.linspace(0,100,100)
plt.plot(x,x)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```
